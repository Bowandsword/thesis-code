```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import pandas as pd
from PIL import Image
import random
from skimage.draw import polygon
import math

# Create the "data Parent" folder if it doesn't exist
parent_folder = 'Yolo_test_shapes'
if not os.path.exists(parent_folder):
    os.makedirs(parent_folder)

# Create the "data Parent" folder if it doesn't exist
train_folder = parent_folder+'/train'
if not os.path.exists(train_folder):
    os.makedirs(train_folder)    
    
# Create the "data Parent" folder if it doesn't exist
val_folder = parent_folder+'/val'
if not os.path.exists(val_folder):
    os.makedirs(val_folder)    
    
    
    
# Create the "images" folder if it doesn't exist
image_train_folder = train_folder+'/images'
if not os.path.exists(image_train_folder):
    os.makedirs(image_train_folder)

# Create the "images" folder if it doesn't exist
label_train_folder = train_folder+'/labels'
if not os.path.exists(label_train_folder):
    os.makedirs(label_train_folder)    
    
# Create the "images" folder if it doesn't exist
image_val_folder = val_folder+'/images'
if not os.path.exists(image_val_folder):
    os.makedirs(image_val_folder)

# Create the "images" folder if it doesn't exist
label_val_folder = val_folder+'/labels'
if not os.path.exists(label_val_folder):
    os.makedirs(label_val_folder)       
    
    
# Initialize a list to store the data
data = []    
    
width = 512
number = 100

# Initialize a base image matrix
image_base = np.zeros((width, width))

i = 1

while i <= number:
    # Reset image_base
    
    image_base = np.zeros((width, width)) + 150
    # Convert to a 3-channel image (RGB)
    image_base_rgb = np.stack((image_base,) * 3, axis=-1)
    image_base_rgb[:, :] = [0, 0, 255]
    
    
    
    mask_base = np.zeros((width, width)) + 1
    
    
    radius = round((np.random.rand()+0.1 ) * width / 5) + 5

    x_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)
    y_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)

    
    
    random_int = random.randint(0, 1)
    #random_int = 1
    if random_int == 0 : # make circle

        # Create a grid of coordinates
        y, x = np.meshgrid(np.arange(1, width+1), np.arange(1, width+1))

        # Calculate the distance of each point from the center
        distance = np.sqrt((y - y_pos) ** 2 + (x - x_pos) ** 2)

        # Set points within the radius to 1
        #mage_base[distance <= radius] = [0,0,255]
        image_base_rgb[distance <= radius]= [255, 0, 0]
        
        
        
        mask_base[distance <= radius] = 2
        shape = 'circle'
        
        area = round(3.14 * radius * radius) 
        
        surface2d = round(radius * 3.14)
        
        volume = round(4/3 * 3.14 * radius * radius * radius) 
    
    # Calculate clamped coordinates for the square
    x_start = max(0, x_pos - round(radius))
    x_end = min(width, x_pos + round(radius))
    y_start = max(0, y_pos - round(radius))
    y_end = min(width, y_pos + round(radius))

    if random_int == 1:  # Square
        if x_start < x_end and y_start < y_end:  # Ensure valid slice
            image_base_rgb[x_start:x_end, y_start:y_end, :] = [255, 0, 0]
            mask_base[x_start:x_end, y_start:y_end] = 2
            shape = 'square'
            area = round(radius * radius)
            surface2d = round(radius * 4)
            volume = round(radius * radius * radius)
             
    
    # Save the image if needed (optional)
    image = Image.fromarray(image_base_rgb.astype(np.uint8)) 
    string = str(random_int) + " " + str(y_pos/width)+" " + str(x_pos/width)+" "+str(radius*2/width) + " " +str(radius*2/width)
   

    trainvsval = random.randint(1, 2)
    if trainvsval == 1 : # train
        name_label = label_train_folder+f'/image_{i}.txt'        
        
        file = open(name_label, 'w')
        file.write(string)
        file.close()
    
        image.save(image_train_folder+f'/image_{i}.png')

    if trainvsval == 2 : # val
        name_label = label_val_folder+f'/image_{i}.txt'        
        
        file = open(name_label, 'w')
        file.write(string)
        file.close()
    
        image.save(image_val_folder+f'/image_{i}.png')    

    

    
    i += 1
    #print(i)
    

```
