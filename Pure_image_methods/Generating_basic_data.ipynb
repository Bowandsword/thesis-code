```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import pandas as pd
from PIL import Image
import random
from skimage.draw import polygon
import math


# Create the "images" folder if it doesn't exist
output_folder = 'images_assorted'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Create the "images" folder if it doesn't exist
mask_folder = 'masks_assorted'
if not os.path.exists(mask_folder):
    os.makedirs(mask_folder)    
    
# Initialize a list to store the data
data = []    
    
width = 512
number = 1000

# Initialize a base image matrix
image_base = np.zeros((width, width))

i = 1

while i <= number:
    # Reset image_base
    
    image_base = np.zeros((width, width)) + 150
    # Convert to a 3-channel image (RGB)
    image_base_rgb = np.stack((image_base,) * 3, axis=-1)
    image_base_rgb[:, :] = [255, 0, 0]
    
    
    
    mask_base = np.zeros((width, width)) + 1
    
    
    radius = round((np.random.rand()+0.1 ) * width / 5) + 5

    x_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)
    y_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)

    
    
    random_int = random.randint(1, 3)
    #random_int = 1
    if random_int == 1 : # make circle

        # Create a grid of coordinates
        x, y = np.meshgrid(np.arange(1, width+1), np.arange(1, width+1))

        # Calculate the distance of each point from the center
        distance = np.sqrt((x - x_pos) ** 2 + (y - y_pos) ** 2)

        # Set points within the radius to 1
        image_base[distance <= radius] = 255
        image_base_rgb[distance <= radius]= [0, 0, 255]
        
        
        
        mask_base[distance <= radius] = 2
        shape = 'circle'
        
        area = round(3.14 * radius * radius) 
        
        surface2d = round(radius * 3.14)
        
        volume = round(4/3 * 3.14 * radius * radius * radius) 
    
    if random_int == 2 : # make square
    
        image_base[x_pos-round(radius/2):x_pos+round(radius/2), y_pos-round(radius/2):y_pos+round(radius/2)] = 255
        image_base_rgb[x_pos-round(radius/2):x_pos+round(radius/2), y_pos-round(radius/2):y_pos+round(radius/2)]= [0, 0, 255]
        mask_base[x_pos-round(radius/2):x_pos+round(radius/2), y_pos-round(radius/2):y_pos+round(radius/2)] = 2
        shape = 'square'
        
        area = round(radius * radius)
        
        surface2d = round(radius * 4)
        
        volume = round(radius * radius * radius)
        
    if random_int == 3 : # make triangle
    
            # Define triangle vertices
        r = np.array([x_pos-round(radius/2), x_pos, x_pos])
        c = np.array([y_pos-round(radius/2), y_pos, y_pos-round(radius/2)])

        # Create a mask for the triangle
        rr, cc = polygon(r, c)

        # Fill the triangle area with white color
        image_base[rr, cc] = 255
        image_base_rgb[rr, cc] = [0, 0, 255]
        mask_base[rr, cc] = 2
        
        shape = 'triange'
        
        area = round(radius * radius /2)
        
        surface2d = round(radius * 2 + math.sqrt(2 * radius * radius) )
        
        volume = round(radius * radius * radius)        
        
        
    
    # Save the image if needed (optional)
    
    image = Image.fromarray(image_base_rgb.astype(np.uint8))
    mask  = Image.fromarray(mask_base.astype(np.uint8))
    
    image.save(f'images_assorted/image_{i}.png')
    mask.save(f'masks_assorted/mask_{i}.png')
    
    #plt.imsave(os.path.join(output_folder, f'image_{i}.png'), image_base, cmap='gray')
    #plt.imsave(os.path.join(mask_folder, f'masks_{i}.png'), mask_base, cmap='gray')
    
    names = f'../images_assorted/image_{i}.png'
    
    masks = f'../masks_assorted/mask_{i}.png'
    
    data.append([radius, x_pos, y_pos, area, surface2d , volume, names, masks, shape])
    i += 1
    #print(i)
    
# Create a DataFrame from the data list
df = pd.DataFrame(data, columns=['Radius', 'X_Position', 'Y_Position', 'Area' ,'Surface2d', 'Volume','image_path','mask_path','shape'])

# Save the DataFrame as a CSV file
df.to_csv('image_data_assorted.csv', index=False)
```
