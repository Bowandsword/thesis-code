```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import pandas as pd
from PIL import Image
import random
from skimage.draw import polygon
import math
import math

# Create the "images" folder if it doesn't exist
output_folder = 'images_assorted'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Create the "images" folder if it doesn't exist
mask_folder = 'masks_assorted'
if not os.path.exists(mask_folder):
    os.makedirs(mask_folder)    
    
# Initialize a list to store the data
data = []    
    
width = 100
number = 500
slices = 100 
gap = 1
# Initialize a base image matrix
image_base = np.zeros((width, width))

i = 1

while i <= number:
    # Reset image_base
    
    image_base = np.zeros((width, width)) + 150
    # Convert to a 3-channel image (RGB)
    image_base_rgb = np.stack((image_base,) * 3, axis=-1)
    image_base_rgb[:, :] = [255, 0, 0]
    
    Red = [255, 0, 0]
    Green = [0,255,0]
    
    mask_base = np.zeros((width, width)) + 1
    radius = round((np.random.rand()+0.1 ) * width / 5) + 5
    
    x_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)
    y_pos = round((np.random.rand() - 0.5) * width * 0.75 + width / 2)
    z_pos = round((np.random.rand() - 0.5) * slices * gap  * 0.75 + slices * gap / 2)
    
    
    random_int = random.randint(1, 2)
    
 
    
    if random_int == 1 : # make sphere 
        volume = round(radius * radius * radius*4/3 * 3.14)
        slicenumber = 0 
        shape = "sphere"
        
        
        while slicenumber < slices : 
            z = slicenumber * gap
            
            x=0
            y=0 
            while x < width : 
                y=0 
                while y < width : 
                    
                    gapx = x_pos - x
                    gapy = y_pos - y
                    gapz = z_pos - z
                    
                    gap_distance = math.sqrt(gapx*gapx + gapy* gapy)
                    gap_distance = math.sqrt(gap_distance*gap_distance + gapz* gapz)
                    
                    if gap_distance < radius : 
                        image_base_rgb[x,y] = Red
                    
                    if gap_distance > radius : 
                        image_base_rgb[x,y] = Green
                        
                        
                    y=y+1
                    
                x=x+1 
            
        
            image = Image.fromarray(image_base_rgb.astype(np.uint8))

            image.save(f'images_assorted/image_{i}_{slicenumber}.png')
            




            volume = round(radius * radius * radius)

            slicenumber = slicenumber + 1 
            #print(slicenumber)
    
    if random_int == 2 : # make cube  
        volume = round(radius * radius * radius)
        shape = "cube"
        slicenumber = 0 
        x = np.arange(round(x_pos - radius/2) , round(x_pos + radius/2))
        y = np.arange(round(y_pos - radius/2) , round(y_pos + radius/2))
        
        
        while slicenumber < slices : 
            z = slicenumber * gap
            
            #set image to green 
            image_base_rgb[:,:] = Green       

            gapz = abs(z_pos - z)
                  
            if abs(gapz) < radius :  
                

                   
                image_base_rgb[x_pos-round(radius/2):x_pos+round(radius/2), y_pos-round(radius/2):y_pos+round(radius/2)] = Red

        
            image = Image.fromarray(image_base_rgb.astype(np.uint8))

            image.save(f'images_assorted/image_{i}_{slicenumber}.png')
            




            

            slicenumber = slicenumber + 1 
            #print(slicenumber)        
    
    names = f'images_assorted/image_{i}_slicenumber.png'
    data.append([radius, x_pos, y_pos,z_pos , volume, names, shape])
    
    # Save the image if needed (optional)
    
    #image = Image.fromarray(image_base_rgb.astype(np.uint8))
    #mask  = Image.fromarray(mask_base.astype(np.uint8))
    
    #image.save(f'images_assorted/image_{i}.png')
    #mask.save(f'masks_assorted/mask_{i}.png')
    
    #plt.imsave(os.path.join(output_folder, f'image_{i}.png'), image_base, cmap='gray')
    #plt.imsave(os.path.join(mask_folder, f'masks_{i}.png'), mask_base, cmap='gray')
    
   
    
    #masks = f'../masks_assorted/mask_{i}.png'
    
    
    i += 1
    print(i)
    
# Create a DataFrame from the data list
df = pd.DataFrame(data, columns=['Radius', 'X_Position', 'Y_Position','Z_Position', 'Volume','image_path','shape'])

# Save the DataFrame as a CSV file
df.to_csv('volume_data_assorted.csv', index=False)
```
